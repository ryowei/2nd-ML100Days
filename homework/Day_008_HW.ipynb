## Merge / Transform

```python
# 生成範例用的資料 ()
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])

df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
```

```python
# 沿縱軸合併
result = pd.concat([df1, df2, df3])
result
```

```python
# 沿橫軸合併
result = pd.concat([df1, df4], axis = 1)
result
```

```python
# 沿橫軸合併
result = pd.concat([df1, df4], axis = 1, join = 'inner') # 硬串接
print(result)
```

```python
# 沿橫軸合併
result = pd.merge(df1, df4, how='inner')
print(result)
```

```python
# 將 欄-列 逐一解開
print(df1)
df1.melt()
```

## Homework

 1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP

    + 0 個小孩
    + 有 1 - 2 個小孩
    + 有 3 - 5 個小孩
    + 有超過 5 個小孩
 2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot
 
 3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 Z 轉換 後的分數

```python
# 1

child = app_train["CNT_CHILDREN"]
child_cut = [-1, 0, 1, 2, 3, 5, np.max(child)]
app_train['CNT_CHILDREN_GROUP'] = pd.cut(app_train["CNT_CHILDREN"], child_cut, include_lowest = False)

app_train['CNT_CHILDREN_GROUP'].value_counts()                                    
    
```

```python
# 2-1

grp = ["CNT_CHILDREN_GROUP","TARGET"]

group_by_result = app_train.groupby(grp)['AMT_INCOME_TOTAL']
group_by_result.mean()
```

```python
# 2-2

plt_column = 'AMT_INCOME_TOTAL'
plt_by = ['CNT_CHILDREN_GROUP','TARGET']

app_train.boxplot(column = plt_column, by = plt_by, showfliers = False, figsize = (15,8))
plt.suptitle('')
plt.show()
```

```python
# 3

app_train['AMT_INCOME_TOTAL_Z_SCORE_BY_CHILDREN_GRP_TARGET'] = group_by_result.apply(lambda x: (x-np.mean(x))/np.std(x))

app_train[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_Z_SCORE_BY_CHILDREN_GRP_TARGET']].head()
```
